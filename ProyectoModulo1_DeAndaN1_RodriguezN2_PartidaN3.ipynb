### 1.5 Visualización de resultados de simulación.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("ticks")
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
import scipy.optimize as opt
```

```python
import pandas as pd
```

```python
data_score = pd.read_csv("prueba.csv")
```

```python
data_score[:]
```

```python
p = input("¿Qué país quieres?")
```

```python
x = data_score["anio"]
y = data_score[p]
```

```python
plt.figure(figsize = (9,9))
plt.scatter(x, y,lw = 0, cmap = "viridis")

plt.show()
```

```python
def obj(b, x, y, n):
    return np.sum((y - np.array([x**i for i in range(n + 1)]).T.dot(b))**2)
```

```python

error = []
plt.figure(figsize = (8,8))
plt.scatter(x, y)

for i in range(1,8):
    b0 = np.random.random((i + 1,))
    res = opt.minimize(obj, b0, args=(x,y, i))
    t = np.linspace(1986, 2070, 27)
    yhat = np.array([t**j for j in range(i + 1)]).T.dot(res.x)
    error.append(np.sum(y - yhat)**2)
    plt.plot(t, yhat, label = "Orden %s" % i)

min = np.min(data_score[p])
plt.legend(bbox_to_anchor=[1, .92])
plt.xlabel("$x$", fontsize = 22)
plt.ylabel("$y$", fontsize = 22)
plt.xlim(xmin = 1980, xmax = 2070)
plt.ylim(ymin = min-5, ymax = 95)
plt.show()
```

```python
error = []
plt.figure(figsize = (8,8))
plt.scatter(x, y)

for i in range(1,8):
    b0 = np.random.random((i + 1,))
    res = opt.minimize(obj, b0, args=(x,y, i))
    t = np.linspace(1986, 2070, 27)
    yhat = np.array([t**j for j in range(i + 1)]).T.dot(res.x)
    error.append(np.sum(y - yhat)**2)

for t in range(4):
    orden = np.where(error == np.min(error))[0][0]
    b0 = np.random.random((orden + 1,))
    res = opt.minimize(obj, b0, args=(x,y, orden))
    t = np.linspace(1986, 2070, 27)
    yhat = np.array([t**j for j in range(orden + 1)]).T.dot(res.x)

plt.plot(t, yhat, label = "Orden %s" % orden)
min = np.min(data_score[p])
plt.legend(bbox_to_anchor=[1, .92])
plt.xlabel("$x$", fontsize = 22)
plt.ylabel("$y$", fontsize = 22)
plt.xlim(xmin = 1980, xmax = 2070)
plt.ylim(ymin = min -5, ymax = 95)
plt.show()
```
